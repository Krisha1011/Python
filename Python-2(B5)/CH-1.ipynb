```python
s=pd.Series([35,70,85,77,87],index=['A','B','C','D','E'],name="Numbers",dtype="float")
print(s)
print()
print(s.values)
print(s.index)
print(s.shape)
print(s.size)
print(s.ndim)
print(s.hasnans)
print(s.name)
print(s.dtypes)
print(s.is_unique)
```

```python
data=[35,70,85,77,87]
s=pd.Series(data,index=['A','B','C','D','E'])
print(s.loc['C'])
print(s.iloc[2])
#print(s.loc[2]) KeyError
#print(s.iloc['C']) TypeError: Cannot index by location index with a non-integer key

print(s.loc['B':'D'])
#print(s.loc[2:4]) TypeError: cannot do slice indexing on Index with these indexers [2] of type int
#print(s.iloc['B':'D']) TypeError: cannot do positional indexing on Index with these indexers [B] of type str
print(s.iloc[2:4])

print(s.iloc[2])
print(s.iloc[4])
```

 loc[]

1.Access labeled index data
2.inbclude end index position

iloc[]

1.Access integer index
2.Exclude end index position

```python
data=[35,70,85,77,87]
s=pd.Series(data,index=['A','B','C','D','E'])

print(s.iloc[[2,4]])
print(s.loc[['C','E']])
```

```python
data=[35,70,85,77,87]
s=pd.Series(data,index=['A','B','C','D','E'])
print(s.loc[s>50])

s.loc['C':'E']=75
print(s)
```
